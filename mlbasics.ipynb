List vs Numpy - Time Taken


```python
from time import process_time
```

Time taken by a list

```python
inbuilt_list=[ i for i in range(1,10000)]
start_time=process_time()
inbuilt_list=[ i +5  for i in range(1,10000)]
end_time=process_time()
print(end_time - start_time)
```

numpy array

```python
np_array=np.array([i for i in range (1,1000)])

start_time = process_time()

np_array = np_array+ 5  

end_time = process_time()

print(end_time - start_time)
```

add a new column


```python
data = {
    "Name": ["Spongebob", "Patrick", "Squidward"],
    "Age": [30, 35, 50]
}
df = pd.DataFrame(data, index = ["e1", "e2", "e3",])
df["job"]=["a1","a2","a3"]

print(df)
```

add a new row

```python
data = {
    "Name": ["Spongebob", "Patrick", "Squidward"],
    "Age": [30, 35, 50]
}
df = pd.DataFrame(data, index = ["e1", "e2", "e3"])
new_row = pd.DataFrame([{ "Name": "alpha", "Age": 33 }],index=["e1"])
df=pd.concat([df,new_row])

print(df)
```
