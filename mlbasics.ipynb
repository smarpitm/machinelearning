List vs Numpy - Time Taken


```python
from time import process_time
```

Time taken by a list

```python
inbuilt_list=[ i for i in range(1,10000)]
start_time=process_time()
inbuilt_list=[ i +5  for i in range(1,10000)]
end_time=process_time()
print(end_time - start_time)
```

numpy array

```python
np_array=np.array([i for i in range (1,1000)])

start_time = process_time()

np_array = np_array+ 5  

end_time = process_time()

print(end_time - start_time)
```

Array Manipulation

```python
array = np.random.randint(0,10,(2,3))
print(array)
print(array.shape)
```
